# Real-Time Emotion Detection using CNN

# Data Preparation

## Objectives

- Load the FER-2013 dataset
- Explore the dataset
- Check class distribution
- Visualize sample images
- Prepare data for CNN training

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import os
import numpy as np

print("TensorFlow Version:", tf.__version__)
```

```python
train_path = "../data/train"
test_path = "../data/test"
```

```python
emotion_classes = sorted(os.listdir(train_path))

print("Emotion Classes:")
print(emotion_classes)
```

```python
print("Total Emotion Classes:", len(emotion_classes))
```

## Number of Images in Each Class

```python
for emotion in emotion_classes:
    count = len(os.listdir(os.path.join(train_path, emotion)))
    print(f"{emotion}: {count} images")
```

## Load Training Dataset

```python
train_dataset = tf.keras.utils.image_dataset_from_directory(
    train_path,
    image_size=(48, 48),
    color_mode="grayscale",
    batch_size=32,
    shuffle=True
)
```

```python
class_names = train_dataset.class_names

print(class_names)
```

## Visualize Sample Images

```python
plt.figure(figsize=(10, 10))

for images, labels in train_dataset.take(1):

    for i in range(9):

        plt.subplot(3, 3, i + 1)

        plt.imshow(images[i].numpy().astype("uint8").squeeze(), cmap="gray")

        plt.title(class_names[labels[i]])

        plt.axis("off")

plt.tight_layout()
plt.show()
```

## Load Testing Dataset

```python
test_dataset = tf.keras.utils.image_dataset_from_directory(
    test_path,
    image_size=(48, 48),
    color_mode="grayscale",
    batch_size=32,
    shuffle=False
)
```

```python
print("Training Batches :", len(train_dataset))
print("Testing Batches  :", len(test_dataset))
```
